**Note:**
- You must not use built-in functions to solve the problem; use only conditional statements.
- You must not use any import statement.

```
# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab
```

````python
# solution cell
def smallest_value(a: float, b: float, c: float) -> float:
    if a <= b and a <= c:
        return a
    if b <= a and b <= c:
        return b
    return c
````

**Examples**

Input: 'test'

Output: 'TEST'

Explanation: Here 'test' is in lower case, so we converted it to upper case.

````python
# solution cell
def reverse_lower_and_upper_case(x: str) -> str:
    if x.islower() is True:
        return x.upper()
    return x.lower()
````

### Task 4

Create a Python function called `mobile_number_validate` that receives a mobile number as a string and validates whether it is a correct mobile number (returning `True`) or not (returning `False`). 

Rules for mobile number:
- string starts with either of `0`, `46`, or `+46`;
- the remaining string has length 9 such that the first digit is between 1 to 9 and the other digits are between 0 to 9.

```
# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab
```

````python
# solution cell
def mobile_number_validate(number: str) -> bool:
    if number[0] == "0":
        rest = number[1:]
    elif number[0:2] == "46":
        rest = number[2:]
    elif number[0:3] == "+46":
        rest = number[3:]
    else:
        return False
    if len(rest) != 9:
        return False
    if rest[0] == "0":
        return False
    numb = 0
    while numb < len(rest):
        if rest[numb] < "0" or rest[numb] > "9":
            return False
        numb += 1
    return True
````

### Task 5

Create a Python function called `sum_multiples` that receives two parameters:

- `n: int`
- `k: int`

The function must compute the sum of all numbers from `1` to `n` (inclusive) that are divisible by `k`.

**Note:**
- If `n` is less than `1`, return `0`.

```
# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab
```

````python
# solution cell
def sum_multiples(n: int, k: int) -> int:
    summa = 0
    for x in range(1, n + 1):
        if x % k == 0:
            summa += x
    return summa
````

### Task 6

Create a Python function called `alternating_sum` that receives one parameter:

- `n: int`

The function must compute the sum of numbers from `1` to `n` (inclusive), where:

- **odd numbers are added**
- **even numbers are subtracted**

If `n` is less than `1`, return `0`.

#### Example

```python
alternating_sum(5) → returns 3
```

Explanation: ```1 - 2 + 3 - 4 + 5 = 3```

```
# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab
```

````python
# solution cell
def alternating_sum(n: int) -> int:
    summa = 0
    for x in range(1, n + 1):
        if x % 2 == 0:
            summa -= x
        else:
            summa += x
    return summa
````
