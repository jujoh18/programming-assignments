### Winter Month Checker

Create a function `is_winter_month(month)` that takes a string `month` (for example, `"January"` or `"december"`) and returns:

- **`True`** if the month is **December**, **January**, or **February**
- **`False`** otherwise

#### Requirements
- The check must be **case-insensitive**.
- Ignore any **leading or trailing whitespace** in the input.

#### Example
- `is_winter_month("January")` → `True`
- `is_winter_month(" July ")` → `False`


```python
# solution cell
def is_winter_month(month: str) -> bool:
    """
    Determine wether a given month belongs to winter.
    Arg:
       month: Month name
    Returns:
       True if month is in winter otherwise False.
    """
    normalized_month = month.strip().lower()
    if normalized_month in ['december', 'January', 'february']:
        return True
    return False
```

### Parity Labeler

Create a function `parity_label(n)` that returns a string describing whether an integer is even or odd:

- **`"even"`** if `n` is divisible by 2
- **`"odd"`** otherwise

#### Example
- `parity_label(10)` → `"even"`
- `parity_label(7)` → `"odd"`
- `parity_label(0)` → `"even"`


```python
# solution cell
def parity_label(n: int) -> str:
    """
    Classify integer by its parity

    Args:
       n: The integer to classify
    Returns:
       "even" if in is even, otherwise "odd".
    """
    if n % 2 == 0:
        return "even"
    return "odd"
```

### PIN Code Validator

Create a function `is_valid_pin(code)` that checks whether a PIN code is valid and returns **`True`** only if **all** of the following conditions are met:

- The PIN has **exactly 4 characters**
- Every character in the PIN is a **digit**

If either condition is not satisfied, the function should return **`False`**.

#### Example
- `is_valid_pin("1234")` → `True`
- `is_valid_pin("12a4")` → `False` (contains a non-digit)
- `is_valid_pin("123")` → `False` (wrong length)


```python
# solution cell
def is_valid_pin(code: str) -> bool:
    """
    Chack wether a PIN code is valid.
    Args:
       code: the PIN code to validate
    Returns:
       True if PIN is valid, False otherwise
    """
    if len(code) != 4:
        return False
    if not code.isdigit():
        return False
    return True
```

### Missing Value Checker

Create a function `is_missing_value(x)` that returns:

- **`True`** if `x` is one of these missing values: **`None`**, **`""`**, or **`[]`**
- **`False`** otherwise

#### Examples
- `is_missing_value(None)` → `True`
- `is_missing_value("")` → `True`
- `is_missing_value([])` → `True`
- `is_missing_value(0)` → `False`
- `is_missing_value({})` → `False`


```python
# solution cell
def is_missing_value(x: object) -> bool:
    """
    Checks wether a value should be treated missing
    A missing value is one of the following:
    - None
    - an empty string
    - an empty list
    Args:
        x: Any value of any type to check
    Returns:
        True if x is considered missing, otherwise False.
    """
    if x is None:
        return True
    if x == "":
        return True
    if x == []:
        return True
    return False
```
