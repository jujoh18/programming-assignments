### Count Up Generator

Create a function `count_up(n)` that returns a list counting up from `1` to `n`.

#### Rules
- If `n = 4`, the function should return:
  `[1, 2, 3, 4]`
- If `n <= 0`, the function should return an empty list:
  `[]`
- You **must use a `while` loop** to implement the function.

#### Example
- `count_up(5)` -> `[1, 2, 3, 4, 5]`
- `count_up(0)` -> `[]`


```python
# solution cell
def count_up(n: int) -> list[int]:
    """
    Create a list counting up from 1 to n.
    Args:
        n: Ending integer.
    returns:
        A list containing [n,...,n].
        if n is less than or equal to 0, an empty list is returned.
    """
    result: list[int] = []
    current = 1
    while current <= n:
        result.append(current)
        current += 1
    return result
```

### First Long Word Finder

Create a function `first_long_word(words)` that returns the **first word with more than 4 characters** from the list `words`.

#### Rules
- If the list contains a word longer than 4 characters, return the **first** one found.
- If there is **no such word**, return `None`.
- You **must use a loop and `break`** to implement the function.

#### Example
- `first_long_word(["cat", "mouse", "dog"])` -> `"mouse"`
- `first_long_word(["sun", "sky"])` -> `None`


```python
# solution cell
def first_long_word(words: list[str]) -> str | None:
    """
    Return the first word longer than 4 characters in a list.
    The function iterates through the list and stop as soon as a long word is found.
    If not such word exists, None is returned.
    Args:
       words: Alist of strings.
    Returns
       The first string in the list with lenght greater than or equal to 4,
       or none if such word exist.
    """
    result = None
    for word in words:
        if len(word) > 4:
            result = word
            break
    return result
```
