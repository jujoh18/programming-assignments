Create a Python function called `group_by_first_letter` that receives a list of strings.

The function must return a dictionary where:
- Each key is the first letter of a word (case-sensitive).
- Each value is a list of words starting with that letter.
- Words must remain in their original order.
- Empty strings must be ignored.

Rules
- If the input list is empty, return an empty dictionary.
- You must use `defaultdict` from `collections` module. Do not manually implement key-initialization logic.

# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab

````python
# solution cell
from collections import defaultdict


def group_by_first_letter(x: list[str]) -> dict[str, list[str]]:
    d = defaultdict(list)
    for word in x:
        if word == "":
            continue
        första = word[0]
        d[första].append(word)
    return dict(d)
````

- All elements from list 1
- Followed by all elements from list 2

#### Rules

- Do not use: `+` or `.extend()` or manual loops

```
# write here your pseudocode (not graded)

```

````python
# solution cell
from itertools import chain


def flatten_sequences(x: list[int], y: list[int]) -> list[int]:
    if x == [] and y == []:
        return []
    return list(chain(x, y))
````

### Task 3

Create a Python function called `gm_median` that receives a list containing elements of `int`, `float`, and `str` types.
The function must return a `tuple` (geometric mean, median) computed using the `statistics` module.

The function must:
- Ignore elements that are not numbers (int or float)
- Ignore numeric values that are less than or equal to 0
- If the original list is empty, return None
- If, after filtering, no valid numeric values remain, return None
- Otherwise, return a tuple (geometric_mean, median) computed using the statistics module

You must import only the required functions. Do not import the entire module or use wildcard imports.

```
# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab
```

````python
# solution cell
from statistics import geometric_mean, median


def gm_median(x: list[int | float | str]) -> tuple[float, float] | None:
    ny: list[float | int] = []
    if not x:
        return None
    for nummer in x:
        if isinstance(nummer, (int, float)) and nummer > 0:
            ny.append(nummer)
    if not ny:
        return None
    return (geometric_mean(ny), median(ny))
````

### Task 4

Create a Python function called `random_mobile_number_generator` that randomly generates a mobile number (`str`).

The number must:
- Start with one of: `"0"`, `"46"`, `"+46"`
- The remaining part must contain exactly 9 digits
- The first digit of the remaining part must be `"3"` or `"7"`
- The remaining 8 digits must be between `0` and `9`

#### Additional requirement

The function must accept an optional parameter: `seed: int | None = None`
- If `seed` is provided, the output must be deterministic.
- Use `random.Random(seed)` to achieve this.

You must import and use module `random` in your solution. You must import only the required functions. Do not import the entire module or use wildcard imports.

```
# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab
```

````python
# solution cell
from random import Random


def random_mobile_number_generator(seed: int | None = None) -> str:
    rng = Random(seed)

    prefix = rng.choice(["0", "46", "+46"])
    first_digit = rng.choice(["3", "7"])
    remaining_digits = "".join(str(rng.randint(0, 9)) for _ in range(8))

    return prefix + first_digit + remaining_digits
````

### Task 5

Create a Python function called `lottery_numbers` that receives three parameters:
1. number: how many numbers should be generated
2. upper: the upper bound (inclusive)
3. lower: the lower bound (inclusive)

The function must generate a list of unique integer numbers with exactly number elements, each between lower and upper (inclusive).

#### Validation rules
Return None if any of the following is true:
- lower < 0
- upper < 1
- number < 1
- (upper - lower) < (number - 1) (not enough unique integers available)

You must import and use module `random`. You must import only the required functions.
Do not import random and do not use wildcard imports.

```
# write here your pseudocode (not graded)

# * pseudocode is necessary when asking for help during the lab
```

````python
# solution cell
from random import Random


def lottery_numbers(number: int, upper: int, lower: int) -> list[int] | None:
    if lower < 0 or upper < 1 or number < 1 or (upper - lower) < (number - 1):
        return None
    rng = Random()
    lista: list[int] = []
    nytt = 0
    while len(lista) < number:
        nytt = rng.randint(lower, upper)
        if nytt not in lista:
            lista.append(nytt)
    return lista
````

### Task 6

Create a Python function called `datetime_validator` that receives a string representing a date-time and converts it into a `datetime` object.

Accepted formats (exact):
1. "January 15, 2022, 12:30:45" (full month name)
2. "Jan 15, 2022, 12:30:45" (abbreviated month name)

Rules:
- Ignore leading/trailing whitespace in the input (e.g., "  January 15, 2022, 12:30:45  " is valid).
- If parsing succeeds in either format, return the `datetime` object.
- If parsing fails, return `None`.

Implementation requirements:
- You must use [`datetime.strptime()`](https://docs.python.org/3/library/datetime.html#datetime.datetime.strptime) and a `try/except` block. Check the [format codes](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes).
- Import requirement (direct style): `from datetime import datetime`
- Do not use import `datetime` and do not use wildcard imports.

````python
# solution cell
from datetime import datetime


def datetime_validator(s: str) -> datetime | None:
    s = s.strip()

    formats = [
        "%B %d, %Y, %H:%M:%S",
        "%b %d, %Y, %H:%M:%S",
    ]

    for fmt in formats:
        try:
            return datetime.strptime(s, fmt)
        except ValueError:
            continue

    return None
````

### Task 7

Create a Python function called `most_common_words` that receives:
1. words: list of strings
2. n: integer

The function must return a list of the n most common words in descending frequency order. Each element in the returned list must be a tuple: (word, count)

#### Rules
- If n < 1, return `None`
- If words is empty, return `None`
- If n is greater than the number of unique words, return all unique words
- Words are case-sensitive
- You must use `Counter` from `collections` module. Do not manually implement counting logic

````python
# solution cell
from collections import Counter


def most_common_words(words: list[str], n: int) -> list[tuple[str, int]] | None:
    if n < 1 or words == []:
        return None
    c = Counter(words)
    return c.most_common(n)
````
