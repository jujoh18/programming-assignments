Create a Python function called `join_words_multiline` that receives three strings `w1`, `w2`, and `w3` and returns a single string in the form:

`"<w1>-<w2>-<w3>"`

The implementation must use a **multi-line statement** (use parentheses so the expression spans multiple lines without using `\`).

```python
# solution cell
def join_words_multiline(w1: str, w2: str, w3: str) -> str:
    result: str = w1 + "-" + w2 + "-" + w3
    return result
```

## Problem 2

Create a Python function called `quote_report` that receives a string `word` and returns a string containing two lines:
1. The word wrapped in **single quotes**
2. The word wrapped in **double quotes**

The result must contain a newline between the two lines.
Example for `word="hello"`:
- first line: `'hello'`
- second line: `"hello"`

```python
# solution cell
def quote_report(word: str) -> str:
    first: str = f"'{word}'"
    second: str = f"\"{word}\""
    return f"{first}\n{second}"
```

## Problem 3

Create a Python function called `case_transformations` that receives a string `text` and returns a tuple with:

1. lower case  
2. upper case  
3. capitalized  
4. title case  

The return type must be `tuple[str, str, str, str]`.

```python
# solution cell
def case_transformations(text: str) -> tuple[str, str, str, str]:
    return (
        text.lower(),
        text.upper(),
        text.capitalize(),
        text.title()
    )
```

## Problem 4

Create a Python function called `safe_concat` that receives a string `prefix` and a value `x` that can be an `int` or a `float`.

The function must return the concatenation of `prefix` and `x`, but **must not** try to concatenate a string and a number directly.
Instead, it must convert the number to a string first.

Use a type hint for `x` that allows both `int` and `float`.

```python
# solution cell
def safe_concat(prefix: str, x: int | float) -> str:
    return prefix + str(x)
```
